```python
import requests
import json
import getpass

DATABRICKS_HOST = "https://dbc-67e4f108-6ec3.cloud.databricks.com"
TOKEN = getpass.getpass("Enter your Databricks PAT: ")

SCOPE_NAME = "nba_secrets"
KEY_NAME = "balldontlie_api_key"
SECRET_VALUE = getpass.getpass(prompt=f"Enter the value for '{KEY_NAME}' in scope '{SCOPE_NAME}': ")

API_ENDPOINT = f"{DATABRICKS_HOST}/api/2.0/secrets/put"
headers = {
    "Authorization": f"Bearer {TOKEN}",
    "Content-Type": "application/json"
}

payload = {
    "scope": SCOPE_NAME,
    "key": KEY_NAME,
    "string_value": SECRET_VALUE
}

print(f"Attempting to put secret '{KEY_NAME}' into scope '{SCOPE_NAME}' at {API_ENDPOINT}...")
response = requests.post(API_ENDPOINT, headers=headers, data=json.dumps(payload))
    
if response.status_code == 200:
    print(f"Success! Secret key '{KEY_NAME}' added to scope '{SCOPE_NAME}'.")
else:
    print(f"Error adding secret. Status Code: {response.status_code}")
    print(f"Error Message: {response.text}")
        
```
